```python
import pandas as pd
import numpy as np
```

```python
employee = pd.read_csv('employee_preprocessed.csv')
attendance = pd.read_csv('attendance_preprocessed.csv')
salary = pd.read_csv('salary_preprocessed.csv')
leaves = pd.read_csv('leaves_preprocessed.csv')
```

```python
# join employee and attendance left join Employee_No
df = pd.merge(employee, attendance, how='left', on='Employee_No')
df.head()
```

```python
# merge df and salary left join Employee_No
df = pd.merge(df, salary, how='left', on='Employee_No')
df.head()
```

```python
#merge df and leaves left join Employee_No
df = pd.merge(df, leaves, how='left', on='Employee_No')
df.head()
```

```python
df.to_csv('employee_merged.csv', index=False)
```
