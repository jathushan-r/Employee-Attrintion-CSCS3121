```python
#change cols in ear_cols to numeric
for col in ear_cols:
    salary[col] = pd.to_numeric(salary[col], errors='coerce')
#change cols in ded_cols to numeric
for col in ded_cols:
    salary[col] = pd.to_numeric(salary[col], errors='coerce')
    
```

```python
# create Total Earning and Total Deduction columns in salary adding all the columns in ear_cols and ded_cols respectively
salary['Total Earnings'] = salary[ear_cols].sum(axis=1)
salary['Total Deductions'] = salary[ded_cols].sum(axis=1)
```

```python
#remove all the columns in ear_cols and ded_cols
salary.drop(ear_cols, axis=1, inplace=True)
salary.drop(ded_cols, axis=1, inplace=True)
salary.head()
```

```python
#remove all cols in unwant_cols
salary.drop(unwant_cols, axis=1, inplace=True)
salary.head()
```

```python
# Net salary = Total Earnings - Total Deductions
salary['Net Salary'] = salary['Total Earnings'] - salary['Total Deductions']
salary.head()
```

```python
#unique values in Area
print(salary['Area'].unique())
#change \\N to NaN
salary['Area'] = salary['Area'].replace('\\N', np.nan)
```
